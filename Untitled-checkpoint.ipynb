```python
import math
import numpy as np
c=np.random.randn(1_000_000)
a=0
for i in c:
    a+=i
mean=a/len(c)


b=0
for i in c:
    b+=(i-mean)**2
stddev=math.sqrt(b/len(c))

print(mean)
print(stddev)
```

```python
import numpy as np
a = np.random.randn(1000,1000)
b=np.mean(a,axis=1,keepdims=True)
c=np.std(a,axis=1,keepdims=True)
print(b)
print(c)
d = (a - b) / c
print("Shape of normalized a:", d.shape)
```
